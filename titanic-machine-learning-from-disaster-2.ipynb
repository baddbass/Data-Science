```python
train = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')
train.head()
```

```python
train['Survived'].value_counts()
```

We see that 74% of Females survived and 81% of Males survived.

Let's convert the Sex column to numerical values. 0 for Male and 1 for Female.

```python
train.replace(to_replace =['male', 'female'], value=[0,1], inplace=True)
```

```python
train.head()
```

```python
train.groupby(['Pclass'])['Survived'].value_counts(normalize=True)
```

From the table we can see majority of first class passengers survived the most compared to 2nd and 3rd class passengers.

**FEATURE SELECTION AND DATA NORMALIZATION**

```python
X = train[['Pclass', 'Sex', 'SibSp', 'Parch']]
X[0:5]
```

```python
y = train['Survived'].values
y[0:5]
```

```python
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

**USING DECISION TREE CLASSIFIER**

```python
#First we split our data into test and train 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)
print('Train set: ', X_train.shape, y_train.shape)
print('Test set: ', X_test.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT_Model = DecisionTreeClassifier(criterion="entropy", max_depth=4)
DT_Model.fit(X_train, y_train)
DT_Model
```

```python
yhat =DT_Model.predict(X_test)
yhat
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(yhat, y_test)
```

**NEXT, WE FIT THE MODEL WITH THE COMPLETE X DATA PROVIDED AND TRY TO PREDICT THE TEST DATA**

```python
DT_model = DecisionTreeClassifier(criterion="entropy", max_depth=4)
DT_model.fit(X, y)
DT_model
```

```python
#Let's assume male to be 0 and female to be 1
test.replace(to_replace=['male', 'female'], value=[0,1], inplace=True)
```

```python
#Let's normalize and format our test data 
XT = test[['Pclass', 'Sex', 'SibSp', 'Parch']]
XT.head()
```

```python
XT = preprocessing.StandardScaler().fit(XT).transform(XT)
XT
```

```python
yhat2 = DT_model.predict(XT)
yhat2
```

```python
df = test[['PassengerId']].reset_index(drop=True)
df['Survived'] = yhat2
df.head()
```

```python
df.to_csv('result_gender_submission.csv', index=False)
```
